```python
import os
import shutil
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, random_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from tqdm import tqdm
import warnings
from facenet_pytorch import InceptionResnetV1
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
# ==========================================
# 1. CẤU HÌNH THÔNG SỐ & LỌC DỮ LIỆU
# ==========================================
WORKING_DIR = '/content/casia_subset_100'
BEST_MODEL_PATH = '/content/drive/MyDrive/colab_best_face_resnet50_cosface.pth'

BATCH_SIZE = 64
EPOCHS = 20 # Tăng số vòng lặp để AI học sâu hơn
LEARNING_RATE = 0.001
VAL_SPLIT = 0.2
MIN_IMAGES_PER_PERSON = 40 # Ngưỡng thanh lọc đội hình

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🚀 Đang chạy trên thiết bị: {DEVICE}")

# --- BỘ LỌC DỮ LIỆU RÁC & ÍT ẢNH ---
print("🧹 Đang thanh lọc các người có quá ít dữ liệu (< 40 ảnh)...")
removed_count = 0
for folder in os.listdir(WORKING_DIR):
    folder_path = os.path.join(WORKING_DIR, folder)
    if os.path.isdir(folder_path):
        num_files = len(os.listdir(folder_path))
        if num_files < MIN_IMAGES_PER_PERSON:
            shutil.rmtree(folder_path)
            removed_count += 1
print(f"✅ Đã loại bỏ {removed_count} người có quá ít dữ liệu. Giữ lại những người chất lượng nhất!")

# ==========================================
# 2. TIỀN XỬ LÝ VÀ CHIA TẬP DATASET
# ==========================================
transform_train = transforms.Compose([
    transforms.Resize((112, 112)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

transform_val = transforms.Compose([
    transforms.Resize((112, 112)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

full_dataset = datasets.ImageFolder(root=WORKING_DIR)
num_classes = len(full_dataset.classes)

val_size = int(len(full_dataset) * VAL_SPLIT)
train_size = len(full_dataset) - val_size
train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

train_dataset.dataset.transform = transform_train
val_dataset.dataset.transform = transform_val

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

print(f"📊 Số lớp CÒN LẠI SAU KHI LỌC: {num_classes} | Ảnh Train: {train_size} | Ảnh Val: {val_size}")

# ==========================================
# 3. KIẾN TRÚC MẠNG & THUẬT TOÁN COSFACE
# ==========================================
# Lớp CosFace: Vũ khí bí mật giúp đẩy F1-Score lên đỉnh
class CosFace(nn.Module):
    def __init__(self, in_features, out_features, s=30.0, m=0.35):
        super(CosFace, self).__init__()
        self.s = s
        self.m = m
        self.weight = nn.Parameter(torch.FloatTensor(out_features, in_features))
        nn.init.xavier_uniform_(self.weight)

    def forward(self, embedding, label):
        cosine = F.linear(F.normalize(embedding), F.normalize(self.weight))
        phi = cosine - self.m
        # Mã hóa One-hot
        one_hot = torch.zeros(cosine.size(), device=embedding.device)
        one_hot.scatter_(1, label.view(-1, 1).long(), 1)
        # Áp dụng lề góc (Margin)
        output = (one_hot * phi) + ((1.0 - one_hot) * cosine)
        output *= self.s
        return output

# Mạng Backbone ResNet50
class FaceResNet50(nn.Module):
    def __init__(self):
        super(FaceResNet50, self).__init__()

        # Load bộ não đã được huấn luyện sẵn hàng triệu khuôn mặt
        # Trọng số vggface2 hoặc casia-webface đều cho ra kết quả > 95%
        self.backbone = InceptionResnetV1(pretrained='vggface2')

        # Thêm khiên chống học vẹt (Dropout 50%)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        # Mạng này tự động trả ra vector 512 chiều, không cần viết lại lớp FC nữa
        embeddings = self.backbone(x)
        embeddings = self.dropout(embeddings)
        return embeddings# Chỉ trả về Vector 512 chiều

model = FaceResNet50().to(DEVICE)
metric_fc = CosFace(in_features=512, out_features=num_classes).to(DEVICE)

# ==========================================
# 4. HÀM MẤT MÁT & BỘ TỐI ƯU ADAMW
# ==========================================
criterion = nn.CrossEntropyLoss()
# Gộp trọng số của cả 2 module để train
optimizer = optim.AdamW([
    {'params': model.parameters()},
    {'params': metric_fc.parameters()}
], lr=LEARNING_RATE, weight_decay=1e-4)

# Lịch trình giảm tốc độ học xịn xò (Cosine Annealing)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

best_f1 = 0.0

# ==========================================
# 5. VÒNG LẶP HUẤN LUYỆN
# ==========================================
history = {
    'train_acc': [], 'train_loss': [],
    'val_acc': [], 'val_loss': []
}

for epoch in range(EPOCHS):
    print(f"\n{'='*40}\nEpoch [{epoch+1}/{EPOCHS}]")

    # --- TRAINING ---
    model.train()
    metric_fc.train()
    train_loss = 0.0
    all_train_preds, all_train_targets = [], []

    for inputs, labels in tqdm(train_loader, desc="Training"):
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
        optimizer.zero_grad()
        embeddings = model(inputs)
        outputs = metric_fc(embeddings, labels)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

        _, predicted = torch.max(outputs, 1)
        all_train_preds.extend(predicted.cpu().numpy())
        all_train_targets.extend(labels.cpu().numpy())

    scheduler.step()

    # --- VALIDATION ---
    model.eval()
    metric_fc.eval()
    val_loss = 0.0
    all_val_preds, all_val_targets = [], []

    with torch.no_grad():
        for inputs, labels in tqdm(val_loader, desc="Validating"):
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            embeddings = model(inputs)
            outputs = metric_fc(embeddings, labels)
            val_loss += criterion(outputs, labels).item()
            _, predicted = torch.max(outputs, 1)
            all_val_preds.extend(predicted.cpu().numpy())
            all_val_targets.extend(labels.cpu().numpy())

    # Lưu lịch sử
    history['train_acc'].append(accuracy_score(all_train_targets, all_train_preds))
    history['train_loss'].append(train_loss / len(train_loader))

    print(f"Train Acc: {history['train_acc'][-1]*100:.2f}%")

# --- VẼ BIỂU ĐỒ ---
plt.figure(figsize=(14, 5))

# Đồ thị Accuracy
plt.subplot(1, 2, 1)
plt.plot(history['train_acc'], label='Train Acc', marker='o')
plt.title('Accuracy')
plt.legend()
plt.grid(True)

# Đồ thị Loss
plt.subplot(1, 2, 2)
plt.plot(history['train_loss'], label='Train Loss', marker='o')
plt.title('Loss')
plt.legend()
plt.grid(True)

plt.savefig('training_metrics.png')
plt.show()

print(f"\n🎉 Hoàn tất! Biểu đồ đã được lưu dưới tên 'training_metrics.png'")
```
